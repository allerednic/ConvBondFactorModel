```python
df = pd.read_csv('./DataSet/Table_1547.csv', encoding='gb2312')
df.rename(columns={'成交日期      ':'TradeDate', '成交时间    ':'TradeTime', '证券代码    ':'secID',
            '委托方向        ':'Direction', '成交数量    ':'Qnty', '成交均价    ':'AvgPrice',
            '成交金额     ':'Amount', '佣金     ':'Commission', '交易规费    ':'Fee', '证券名称    ':'secName',
            }, inplace=True)
df.head()
```

```python
columns=['secID', 'datetime', 'Price', 'Qnty', 'Commission']
```

```python
df_list = []
for i in range(len(df)):
    if u'转债' in df.loc[i,'secName']:
        tmp_row = {}
        trade_time = pd.to_datetime(df['TradeDate'][i] + df['TradeTime'][i])
        secID = re.findall(r'(\d{6})', df.loc[i, 'secID'])[0]
        tmp_row['secID'] = secID
        tmp_row['datetime'] = trade_time
        tmp_row['Price'] = df.loc[i, 'AvgPrice']
        sign = -1 if '卖出' in df.loc[i, 'Direction'] else 1
        tmp_row['Qnty'] = df.loc[i, 'Qnty'] * sign
        tmp_row['Commission'] = df.loc[i, 'Commission'] + df.loc[i, 'Fee']
        df_list.append(pd.DataFrame(tmp_row, index=[0]))
```

```python
logging = pd.concat(df_list, ignore_index=True)
logging.tail()
```
